```python
import pandas as pd
import numpy as np
import seaborn as sns
from random import uniform
import matplotlib.pyplot as plt
```

```python
def esta_dentro_rombo(x, y, centro_x, centro_y, lado):
    hipotenusa = (lado*lado*2)**(0.5)
    
    return (abs(x-centro_x) + abs(y-centro_y)) <= (hipotenusa/2)
```

#### Creo puntos en el cuadrado que encierra el rombo, si caen dentro del rombo los acepto si no vuelvo a generar

```python
def rombo_gen(n, centro_x, centro_y, lado):
    """Devuelve una lista con n muestras de la distribucion pedida"""
    muestras = []
    hip = (lado*lado*2)**(0.5)
    
    while len(muestras) < n:
        # Me creo puntos x,y que esten en el cuadrado de lado hip que encierra al rombo
        x = uniform(0,1)*hip + centro_x - hip/2
        y = uniform(0,1)*hip + centro_y - hip/2
        if esta_dentro_rombo(x,y, centro_x, centro_y, lado):
            muestras.append((x,y))
            
    return muestras
```

```python
prueba = rombo_gen(10000, 10, 10, 5)
```

```python
df_rombo = pd.DataFrame(prueba, columns=["x", "y"])
```

```python
sns.jointplot(x=df_rombo["x"], y=df_rombo["y"], kind='scatter', s=5)
plt.subplots_adjust(top=0.9)
plt.suptitle('Distribucion uniforme 2D con forma de rombo', fontsize = 20);
```
