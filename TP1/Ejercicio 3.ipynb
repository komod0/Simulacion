```python
def dens_normal(sigma, mu, x):
    a = 1/(sigma*np.sqrt(2*np.pi))
    b = np.exp(-((x-mu)**2)/(2*sigma**2))
    return a*b

def dens_exp(lbda, x):
    return lbda * 1*np.exp(-lbda * abs(x))
```

```python
#Para aceptar o rechazar
c = 75/6

def aceptar(x):
    u = np.random.rand()
    a = dens_normal(3, 15, x)/ (c * (dens_exp(1/15, x)))
    return u < a
    
    
```

```python
cant_gen = 0
rechazados = 0
gen = []
while (cant_gen < 100000):
    x = np.random.exponential(scale = 15, size = 1)[0] #scale = 1/lambda
    u = np.random.rand()
    if (u < 0.5):
        x = -x
    if (aceptar(x)):
        cant_gen+=1
        gen.append(x)
    else: 
        rechazados+=1
```

```python
plt.figure(figsize=[15,10])
sb.distplot(gen, kde = True);
```
