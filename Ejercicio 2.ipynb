```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.set()
```

```python
l = 1/15
```

## Ejercicio a

```python
def f(t):
    if (t>=0 and t<10):
        return 1/25
    if (t>=10):
        return (3/5)*l*(np.e**(-l*(t-10)))
```

```python
dom_a = np.linspace(0,120,1000)
```

```python
img_a = []
for i in (dom_a):
    img_a.append(f(i))
```

```python
fig,ax=plt.subplots(figsize=(10,5))
sns.lineplot(x=dom_a,y=img_a)
plt.xlabel('t')
plt.ylabel('f(t)')
plt.show()
```

## Ejercicio b

```python
def F(t):
    if t <= 10:
        return 1/25 * t
    if t > 10:
        return (3/5)*(1-np.e**(-l*(t-10)))+10/25
    
```

```python
dom_b = np.linspace(0,70,1000)
img_b = []
for i in dom_b:
    img_b.append(F(i))
```

```python
sns.lineplot(x=dom_b,y=img_b)
plt.title('F(t)')
```

```python
def F_inv(u):
    if u < 0.4:
        return 25*u
    if u >= 0.4:
        return -np.log(((2/3)-(5/3)*u)+1)/l + 10
```

```python
dom_b2 = np.linspace(0,0.99,100)
img_b2 = []
for i in dom_b2: 
    img_b2.append(F_inv(i))
```

```python
sns.lineplot(x=dom_b2,y=img_b2)
plt.title('F inversa')
```
