```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
SEED = np.floor((100680+101046+101018+100490) / 4)
```

```python
##Generador uiforme [0,1] del ejercicio (1)
def uniform (n,seed,a,c,m):
    x = seed
    vector = []
    for i in range(n):
        x = ((a*x+c)%m)
        vector.append(x/m)
    return vector
```

```python
#Generamos la funcion de probabilidad acumulada
y = np.array([0.1, 0.6, 0.9, 1]); 
x = np.array(['a', 'b', 'c', 'd'])

plt.figure(figsize = [10,8])
plt.bar(x,y, color = ["red", "blue", "orange", "green"], width = 1, align = 'edge')
plt.title("Funcion de probabilidad acumulada", fontsize = 20)
plt.grid(True)
```

```python
rand = uniform(100000,SEED,1013904223,1664525,2**32)
gen = []
for num in rand:
    if (num <= 0.1):
        gen.append(0.1)
    elif ((num > 0.1) & (num <= 0.6)):
        gen.append(0.5)
    elif ((num > 0.6) & (num <= 0.9)):
        gen.append(0.3)
    else:
        gen.append(0.1)
```

```python
plt.figure(figsize = [10,8])
plt.hist(gen, color = ["orange"], width = 0.1, rwidth = 1000, align = "mid")
plt.title("Histograma numeros generados", fontsize = 20)
plt.xticks([0.1, 0.3, 0.5])
plt.grid(True)
```
