```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from random import uniform
from math import sqrt, log, cos, sin, pi
from scipy.stats import norm
```

#### Transformada de Box-Mueller:

$(1)\ X = \sqrt{-2ln(U_{1})}cos(2\pi U_{2})$

$(2)\ Y = \sqrt{-2ln(U_{1})}sin(2\pi U_{2})$

_Sacado del libro Introduction to probability models - Ross, Pag. 682_

```python
def box_mueller_1(n):
    muestras = []

    for i in range(n):
        u1 = uniform(0,1)
        u2 = uniform(0,1)
        x = sqrt(-2*log(u1))*cos(2*pi*u2)
        muestras.append(x)
    return muestras
```

```python
def box_mueller_2(n):
    muestras = []
    
    for i in range(n):
        u1 = uniform(0,1)
        u2 = uniform(0,1)
        x = sqrt(-2*log(u1))*sin(2*pi*u2)
        muestras.append(x)
    return muestras
```

```python
prueba = box_mueller_1(100000)
plt.figure(figsize=(15,8))
plt.hist(prueba, bins=50)
plt.title("Transformada de Box-Mueller version 1", fontsize=20);
```

---

```python
prueba = box_mueller_2(100000)
plt.figure(figsize=(15,8))
plt.hist(prueba, bins=40)
plt.title("Transformada de Box-Mueller version 2", fontsize=20);
```

---

### Comparamos la normal con las muestras generadas por Box-Muller 1

```python
muestras_bm_1 = box_mueller_1(100000)
plt.figure(figsize=(15,8))
sns.distplot(muestras_bm_1, fit=norm, fit_kws={"color":"red"})
plt.legend(["Distribucion normal"], fontsize=16)
plt.title("Variable generada por Box-Mueller 1 Vs Distribucion normal", fontsize=20);
```

#### Calculamos la media

```python
np.mean(muestras_bm_1)
```

#### Y la varianza

```python
np.var(muestras_bm_1)
```

---

### Comparamos la normal con las muestras generadas por Box-Muller 2

```python
muestras_bm_2 = box_mueller_2(100000)
plt.figure(figsize=(15,8))
sns.distplot(muestras_bm_2, fit=norm, fit_kws={"color":"red"})
plt.legend(["Distribucion normal"], fontsize=16)
plt.title("Variable generada por Box-Mueller 2 Vs Distribucion normal", fontsize=20);
```

#### Calculamos la media

```python
np.mean(muestras_bm_2)
```

#### Y la varianza

```python
np.var(muestras_bm_2)
```
